```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
```

```python
df = pd.DataFrame({'x': np.arange(0, 10), 'f': [0.0,0.05, 0.1, 0.1, 0.1, 0.15, 0.15, 0.15, 0.1, 0.1]})
df['F'] = df['f'].cumsum()
print(df)
```

```python
#create a function that returns a random number according to the empirical distribution of dataframe df
def random_number(df):
    r = np.random.rand()
    return df.loc[df['F'] >= r, 'x'].iloc[0]
```

```python
#vamos a generar 10000 observaciones
n = 10000
observations = [random_number(df) for i in range(n)]
print(observations)
K=100
a=25
h=2
pv=60
```

```python
def simular(variables,serie , K, a, h, pv):
    s, S = variables
    stockFinal=S
    ingresos=0
    unidadesCompradas=0
    for i in range(len(serie)):
        stockInicial=stockFinal+unidadesCompradas
        ventas=min(serie[i],stockInicial)
        ingresos+=pv*ventas
        stockFinal=stockInicial-ventas
        ingresos-=h*stockFinal
        if stockFinal<=s:
            unidadesCompradas=S-stockFinal
            ingresos -= (K+a*S)
        else:
            unidadesCompradas=0
    #print(s, S, ingresos)
    return ingresos
```

```python
bestI=0
bestJ=0
bestIngresos=0
for i in range(1,10):
    for j in range(10,30):
        ingresos=simular([i,j],observations,K,a,h,pv)
        #print(i,j,ingresos)
        if ingresos>bestIngresos:
            bestIngresos=ingresos
            bestI=i
            bestJ=j
print("bestI:",bestI)
print("bestJ:",bestJ)
print("bestIngresos:",bestIngresos)
```
